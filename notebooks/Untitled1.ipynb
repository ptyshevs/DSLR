```python
import pandas as pd
df = pd.read_csv('resources/dataset_train.csv', index_col=0)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_histogram(df, feature):
    plt.figure(figsize=(8, 6))
    df.groupby('Hogwarts House')[feature].plot(kind='hist', alpha=.5);
    plt.legend();
    plt.title(feature)
    plt.show()
```

```python
courses = ['Arithmancy',
 'Astronomy',
 'Herbology',
 'Defense Against the Dark Arts',
 'Divination',
 'Muggle Studies',
 'Ancient Runes',
 'History of Magic',
 'Transfiguration',
 'Potions',
 'Care of Magical Creatures',
 'Charms',
 'Flying']
```

```python
for course in courses:
    plot_histogram(df, course)
```
